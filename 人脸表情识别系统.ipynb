```python
from PyQt5.QtGui import *
from PyQt5.QtWidgets import *
from PyQt5.QtCore import *
from tensorflow.keras import models
from PyQt5 import QtCore, QtGui, QtWidgets
import numpy as np
import cv2
import sys

# 表情标签
emotion_map = {0: 'Angry', 1: 'Disgust', 2: 'Fear', 3: 'Happy', 4: 'Sad', 5: 'Surprise', 6: 'Neutral'}
# 加载分类器
classifier = cv2.CascadeClassifier(r"\haarcascade_frontalface_default.xml")

class Ui_MainWindow(object):
    def setupUi(self, MainWindow):
        MainWindow.setObjectName("MainWindow")
        MainWindow.resize(813, 578)
        MainWindow.setMinimumSize(QtCore.QSize(813, 578))
        MainWindow.setMaximumSize(QtCore.QSize(813, 578))
        font = QtGui.QFont()
        font.setFamily("PingFang SC")
        MainWindow.setFont(font)
        self.centralwidget = QtWidgets.QWidget(MainWindow)
        self.centralwidget.setFocusPolicy(QtCore.Qt.NoFocus)
        self.centralwidget.setObjectName("centralwidget")
        self.verticalLayoutWidget_2 = QtWidgets.QWidget(self.centralwidget)
        self.verticalLayoutWidget_2.setGeometry(QtCore.QRect(600, 160, 201, 305))
        self.verticalLayoutWidget_2.setObjectName("verticalLayoutWidget_2")
        self.verticalLayout = QtWidgets.QVBoxLayout(self.verticalLayoutWidget_2)
        self.verticalLayout.setContentsMargins(0, 11, 0, 0)
        self.verticalLayout.setSpacing(15)
        self.verticalLayout.setObjectName("verticalLayout")
        self.angryLabel = QtWidgets.QLabel(self.verticalLayoutWidget_2)
        font = QtGui.QFont()
        font.setFamily("PingFang SC")
        font.setPointSize(15)
        self.angryLabel.setFont(font)
        self.angryLabel.setObjectName("angryLabel")
        self.verticalLayout.addWidget(self.angryLabel)
        self.disgustLabel = QtWidgets.QLabel(self.verticalLayoutWidget_2)
        font = QtGui.QFont()
        font.setFamily("PingFang SC")
        font.setPointSize(15)
        self.disgustLabel.setFont(font)
        self.disgustLabel.setObjectName("disgustLabel")
        self.verticalLayout.addWidget(self.disgustLabel)
        self.fearLabel = QtWidgets.QLabel(self.verticalLayoutWidget_2)
        font = QtGui.QFont()
        font.setFamily("PingFang SC")
        font.setPointSize(15)
        self.fearLabel.setFont(font)
        self.fearLabel.setObjectName("fearLabel")
        self.verticalLayout.addWidget(self.fearLabel)
        self.happyLabel = QtWidgets.QLabel(self.verticalLayoutWidget_2)
        font = QtGui.QFont()
        font.setFamily("PingFang SC")
        font.setPointSize(15)
        self.happyLabel.setFont(font)
        self.happyLabel.setObjectName("happyLabel")
        self.verticalLayout.addWidget(self.happyLabel)
        self.sadLabel = QtWidgets.QLabel(self.verticalLayoutWidget_2)
        font = QtGui.QFont()
        font.setFamily("PingFang SC")
        font.setPointSize(15)
        self.sadLabel.setFont(font)
        self.sadLabel.setObjectName("sadLabel")
        self.verticalLayout.addWidget(self.sadLabel)
        self.surpriseLabel = QtWidgets.QLabel(self.verticalLayoutWidget_2)
        font = QtGui.QFont()
        font.setFamily("PingFang SC")
        font.setPointSize(15)
        self.surpriseLabel.setFont(font)
        self.surpriseLabel.setObjectName("surpriseLabel")
        self.verticalLayout.addWidget(self.surpriseLabel)
        self.neutralLabel = QtWidgets.QLabel(self.verticalLayoutWidget_2)
        font = QtGui.QFont()
        font.setFamily("PingFang SC")
        font.setPointSize(15)
        self.neutralLabel.setFont(font)
        self.neutralLabel.setObjectName("neutralLabel")
        self.verticalLayout.addWidget(self.neutralLabel)
        self.layoutWidget = QtWidgets.QWidget(self.centralwidget)
        self.layoutWidget.setGeometry(QtCore.QRect(12, 12, 798, 114))
        self.layoutWidget.setObjectName("layoutWidget")
        self.gridLayout = QtWidgets.QGridLayout(self.layoutWidget)
        self.gridLayout.setContentsMargins(0, 0, 0, 0)
        self.gridLayout.setObjectName("gridLayout")
        self.horizontalLayout = QtWidgets.QHBoxLayout()
        self.horizontalLayout.setObjectName("horizontalLayout")
        self.titleLabel = QtWidgets.QLabel(self.layoutWidget)
        font = QtGui.QFont()
        font.setFamily("PingFang SC")
        font.setPointSize(24)
        font.setBold(False)
        font.setWeight(50)
        self.titleLabel.setFont(font)
        self.titleLabel.setObjectName("titleLabel")
        self.horizontalLayout.addWidget(self.titleLabel)
        self.gridLayout.addLayout(self.horizontalLayout, 0, 0, 1, 1)
        self.horizontalLayout_2 = QtWidgets.QHBoxLayout()
        self.horizontalLayout_2.setObjectName("horizontalLayout_2")
        self.useCameraButton = QtWidgets.QPushButton(self.layoutWidget)
        self.useCameraButton.setEnabled(True)
        font = QtGui.QFont()
        font.setFamily(".PingFang SC")
        font.setPointSize(20)
        font.setBold(False)
        font.setWeight(50)
        self.useCameraButton.setFont(font)
        self.useCameraButton.setObjectName("useCameraButton")
        self.horizontalLayout_2.addWidget(self.useCameraButton)
        self.useImageButton = QtWidgets.QPushButton(self.layoutWidget)
        font = QtGui.QFont()
        font.setFamily(".PingFang SC")
        font.setPointSize(20)
        self.useImageButton.setFont(font)
        self.useImageButton.setObjectName("useImageButton")
        self.horizontalLayout_2.addWidget(self.useImageButton)
        self.useVideoButton = QtWidgets.QPushButton(self.layoutWidget)
        font = QtGui.QFont()
        font.setFamily(".PingFang SC")
        font.setPointSize(20)
        self.useVideoButton.setFont(font)
        self.useVideoButton.setObjectName("useVideoButton")
        self.horizontalLayout_2.addWidget(self.useVideoButton)
        self.selectModelButton = QtWidgets.QPushButton(self.layoutWidget)  # 初始化选择模型按钮
        font = QtGui.QFont()
        font.setFamily(".PingFang SC")
        font.setPointSize(20)
        self.selectModelButton.setFont(font)
        self.selectModelButton.setFocusPolicy(QtCore.Qt.NoFocus)
        self.selectModelButton.setObjectName("selectModelButton")  # 设置对象名称
        self.horizontalLayout_2.addWidget(self.selectModelButton)
        self.gridLayout.addLayout(self.horizontalLayout_2, 1, 0, 1, 1)
        self.videoLabel = QtWidgets.QLabel(self.centralwidget)
        self.videoLabel.setGeometry(QtCore.QRect(10, 140, 561, 371))
        font = QtGui.QFont()
        font.setFamily("PingFang SC")
        font.setPointSize(30)
        self.videoLabel.setFont(font)
        self.videoLabel.setAlignment(QtCore.Qt.AlignCenter)
        self.videoLabel.setObjectName("videoLabel")
        self.label = QtWidgets.QLabel(self.centralwidget)
        self.label.setGeometry(QtCore.QRect(610, 460, 185, 60))
        self.label.setText("")
        self.label.setPixmap(QtGui.QPixmap(":/img/logo.png"))
        self.label.setScaledContents(True)
        self.label.setObjectName("label")
        MainWindow.setCentralWidget(self.centralwidget)
        self.menubar = QtWidgets.QMenuBar(MainWindow)
        self.menubar.setGeometry(QtCore.QRect(0, 0, 813, 21))
        self.menubar.setObjectName("menubar")
        MainWindow.setMenuBar(self.menubar)
        self.statusbar = QtWidgets.QStatusBar(MainWindow)
        self.statusbar.setObjectName("statusbar")
        MainWindow.setStatusBar(self.statusbar)

        self.retranslateUi(MainWindow)
        QtCore.QMetaObject.connectSlotsByName(MainWindow)

    def retranslateUi(self, MainWindow):
        _translate = QtCore.QCoreApplication.translate
        MainWindow.setWindowTitle(_translate("MainWindow", "人脸表情识别系统"))
        self.angryLabel.setText(_translate("MainWindow", "<html><head/><body><p align=\"center\">生气：0.00%</p></body></html>"))
        self.disgustLabel.setText(_translate("MainWindow", "<html><head/><body><p align=\"center\">厌恶：0.00%</p></body></html>"))
        self.fearLabel.setText(_translate("MainWindow", "<html><head/><body><p align=\"center\">害怕：0.00%</p></body></html>"))
        self.happyLabel.setText(_translate("MainWindow", "<html><head/><body><p align=\"center\">高兴：0.00%</p></body></html>"))
        self.sadLabel.setText(_translate("MainWindow", "<html><head/><body><p align=\"center\">悲伤：0.00%</p></body></html>"))
        self.surpriseLabel.setText(_translate("MainWindow", "<html><head/><body><p align=\"center\">惊讶：0.00%</p></body></html>"))
        self.neutralLabel.setText(_translate("MainWindow", "<html><head/><body><p align=\"center\">自然：0.00%</p></body></html>"))
        self.titleLabel.setText(_translate("MainWindow", "<html><head/><body><p align=\"center\">人脸表情识别系统</p></body></html>"))
        self.useCameraButton.setText(_translate("MainWindow", "使用摄像头"))
        self.useImageButton.setText(_translate("MainWindow", "选择图片"))
        self.useVideoButton.setText(_translate("MainWindow", "选择视频"))
        self.selectModelButton.setText(_translate("MainWindow", "选择模型"))  # 设置按钮文本
        self.videoLabel.setText(_translate("MainWindow", "显示区"))

class VideoPageWindow(QMainWindow, Ui_MainWindow):
    def __init__(self, parent=None):
        super(VideoPageWindow, self).__init__(parent)
        self.timer_camera = QTimer()
        self.timer_video = QTimer()
        self.cap = cv2.VideoCapture()
        self.CAM_NUM = 0
        self.setupUi(self)
        self.slot_init()
        self.once = True
        self.model = None  # 初始化模型为None

    def slot_init(self):
        self.timer_camera.timeout.connect(self.show_camera)
        self.timer_video.timeout.connect(self.show_video)
        self.useCameraButton.clicked.connect(self.slotCameraButton)
        self.useVideoButton.clicked.connect(self.slotVideoButton)
        self.useImageButton.clicked.connect(self.slotImageButton)
        self.selectModelButton.clicked.connect(self.slotChooseModelButton)  # 选择模型按钮的连接

    def slotChooseModelButton(self):
        model_path, _ = QFileDialog.getOpenFileName(self, "选择模型文件", "/Models", "模型文件 (*.h5 *.hdf5)")
        if model_path:
            self.model = models.load_model(model_path)
            QMessageBox.information(self, "模型加载成功", f"已加载模型：{model_path}")

    def show_camera(self):
        flag, frame = self.cap.read()  # 捕获帧
        self.image = cv2.flip(frame, 1)  # 镜像翻转
        self.predict(self.image)
        show = cv2.resize(self.image, (480, 320))  # 重置大小
        show = cv2.cvtColor(show, cv2.COLOR_BGR2RGB)  # 转换颜色空间
        showImage = QImage(show.data, show.shape[1], show.shape[0], QImage.Format_RGB888)  # 转换成QImage
        self.videoLabel.setPixmap(QPixmap.fromImage(showImage))  # 设置UI

    def show_video(self):
        flag, self.image = self.cap.read()  # 捕获帧
        if flag:
            self.predict(self.image)
            show = cv2.resize(self.image, (480, 320))  # 重置大小
            show = cv2.cvtColor(show, cv2.COLOR_BGR2RGB)  # 转换颜色空间
            showImage = QImage(show.data, show.shape[1], show.shape[0], QImage.Format_RGB888)  # 转换成QImage
            self.videoLabel.setPixmap(QPixmap.fromImage(showImage))  # 设置UI
    
    def predict(self, image):
        if self.model is None:  # 检查是否已经加载了模型
            return
        # 转换成灰度图像
        gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
        # 检测人脸区域
        faceRects = classifier.detectMultiScale(gray, scaleFactor=1.2, minNeighbors=3, minSize=(32, 32))
        color = (0, 0, 255)

        if len(faceRects) > 0:  # 大于0则检测到人脸
            for faceRect in faceRects:
                x, y, w, h = faceRect         
                # 框出人脸
                cv2.rectangle(image, (x, y), (x + w, y + h), color, 2)
                # 获取人脸源
                src = gray[y:y + h, x:x + w]
                # 检查是否获取到了有效的人脸区域
                if src.size == 0:
                    continue
                # 缩放至48*48
                try:
                    img = cv2.resize(src, (48, 48))
                except cv2.error as e:
                    print(f"Error resizing image: {e}")
                    continue

                # 归一化
                img = img / 255.0
            
                # 扩展维度
                x_input = np.expand_dims(img, axis=0)
                x_input = np.array(x_input, dtype='float32').reshape(-1, 48, 48, 1)
            
                # 预测输出
                y_output = self.model.predict(x_input)
                percent = self.getPercent(y_output[0])
            
                # 更新UI
                self.angryLabel.setText('生气：{:.2%}'.format(percent[0]))
                self.disgustLabel.setText('厌恶：{:.2%}'.format(percent[1]))
                self.fearLabel.setText('害怕：{:.2%}'.format(percent[2]))
                self.happyLabel.setText('高兴：{:.2%}'.format(percent[3]))
                self.sadLabel.setText('悲伤：{:.2%}'.format(percent[4]))
                self.surpriseLabel.setText('惊讶：{:.2%}'.format(percent[5]))
                self.neutralLabel.setText('自然：{:.2%}'.format(percent[6]))
            
                # 获取预测的类别
                output_class = np.argmax(y_output[0])
                cv2.putText(image, emotion_map[output_class], (x, y - 10), cv2.FONT_HERSHEY_COMPLEX, 0.9, (0, 0, 250), 2)

    def getPercent(self, arr):
        sum = np.sum(arr)
        if sum == 0:
            return arr
        return arr / sum


    def slotCameraButton(self):
        if self.timer_camera.isActive() == False:
            self.CAM_NUM = 0
            self.openCamera()
        else:
            self.closeCamera()

    def slotVideoButton(self):
        self.CAM_NUM = QFileDialog.getOpenFileName(self, "选择文件", r"\shiping","视频文件 (*.mp4)")
        if self.CAM_NUM[0]:
            self.CAM_NUM = self.CAM_NUM[0]
            self.openVideo()

    def slotImageButton(self):
        self.CAM_NUM = QFileDialog.getOpenFileName(self, "选择文件",r"\tupian","图像文件 (*.jpg *.jpeg *.bmp *.png)")
        if self.CAM_NUM[0]:
            self.CAM_NUM = self.CAM_NUM[0]
            self.openImage()
    
    

    def openCamera(self):
        if self.timer_video.isActive():
            self.closeVideo()
        flag = self.cap.open(self.CAM_NUM)
        if flag == False:
            msg = QMessageBox.warning(self, u'Warning', u'请监测相机与电脑是否连接正确',
                                      buttons=QMessageBox.Ok,
                                      defaultButton=QMessageBox.Ok)
        else:
            self.timer_camera.start(30)
            self.useCameraButton.setText("关闭摄像头")

    def closeCamera(self):
        self.timer_camera.stop()
        self.cap.release()
        self.videoLabel.clear()
        self.videoLabel.setText("显示区")
        self.useCameraButton.setText("使用摄像头")

    def openVideo(self):
        if self.timer_camera.isActive():
            self.closeCamera()
        flag = self.cap.open(self.CAM_NUM)
        if flag == False:
            msg = QMessageBox.warning(self, u'Warning', u'请检查视频文件是否没有损坏',
                                      buttons=QMessageBox.Ok,
                                      defaultButton=QMessageBox.Ok)
        else:
            self.timer_video.start(30)

    def closeVideo(self):
        self.timer_video.stop()
        self.cap.release()
   
    
    def openImage(self):
        if self.timer_video.isActive():
            self.closeVideo()
        if self.timer_camera.isActive():
            self.closeCamera()
        self.image = cv2.imread(self.CAM_NUM)
        
        # 检查图像是否读取成功
        if self.image is None:
            QMessageBox.warning(self, "警告", "无法读取图像文件，请检查文件路径是否正确。")
            return

        self.predict(self.image)
        
        # 确保图片尺寸不会导致除以0错误
        if self.image.shape[0] == 0 or self.image.shape[1] == 0:
            QMessageBox.warning(self, "警告", "图像文件损坏或格式不正确。")
            return
        # 重置大小并显示图像
        show = cv2.resize(self.image, (int(320 * self.image.shape[1] / self.image.shape[0]), 320))
        show = cv2.cvtColor(show, cv2.COLOR_BGR2RGB)  # 转换颜色空间
        showImage = QImage(show.data, show.shape[1], show.shape[0], QImage.Format_RGB888)  # 转换成QImage
        self.videoLabel.setPixmap(QPixmap.fromImage(showImage))  # 设置UI
        
    def getPercent(self, arr):
        sum = np.sum(arr)
        return arr / sum

if __name__ == "__main__":
    app = QApplication(sys.argv)
    myWin = VideoPageWindow()
    myWin.show()
    sys.exit(app.exec_())
```
